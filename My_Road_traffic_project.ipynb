```python
columns_to_drop = ['fitness_of_casuality', 'casualty_severity', 'age_band_of_casualty','sex_of_casualty','road_allignment'
                   ,'vehicle_driver_relation','owner_of_vehicle','lanes_or_medians','pedestrian_movement',] #based on the graphs show above these cols seemed unnecessary to

# Drop the specified columns from the DataFrame
df3=df2.drop(columns=columns_to_drop)
```

```python
# Filter out categorical columns (object dtype)
categorical_cols = df3.select_dtypes(include=['object']).columns.drop('accident_severity')

# Filter out numerical columns (float and int dtype), excluding the target feature
numerical_cols = df3.select_dtypes(include=['int64']).columns

# Print the list of categorical and numerical columns
print("Categorical Columns:")
print(categorical_cols)
print("\nNumerical Columns:")
print(numerical_cols)
```

```python
# Perform one-hot encoding
df3_encoded = pd.get_dummies(df3, columns=categorical_cols, drop_first=True)
```

```python
from imblearn.over_sampling import SMOTE

X = df3_encoded.drop(columns=['accident_severity'])
y = df3_encoded['accident_severity']

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Display a pie chart to show the balance after SMOTE
plt.figure(figsize=(4, 4))
plt.pie(y_resampled.value_counts(), labels=y_resampled.value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Target Variable after SMOTE')
plt.show()
```

```python
print(df3_encoded.head(5))
```

```python
from sklearn.feature_selection import SelectKBest, chi2  #it wasnt giving me a good idea so i decided to not go with this method
# X_resampled, y_resampled

# Assuming df3 is your dataframe with one-hot encoded categorical features and target
X_encoded = X_resampled  # Features
y_encoded = y_resampled  # Target variable
k=70

# Create SelectKBest object with chi-square test
selector = SelectKBest(score_func=chi2, k=k)

# Fit selector to features and target
X_selected = selector.fit_transform(X_encoded, y_encoded)

# Get the selected feature indices
selected_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_features = X_encoded.columns[selected_indices]

# Print the selected feature names
print("Selected features:", selected_features)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode the target variable y
y_encoded = label_encoder.fit_transform(y)

# Now, y_encoded contains integer labels (0, 1, 2) instead of string labels
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from xgboost import XGBClassifier
from sklearn.svm import SVC
y=y_encoded
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Train and evaluate Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train, y_train)
rf_predictions = rf_classifier.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_predictions)
rf_precision = precision_score(y_test, rf_predictions, average='weighted')
rf_recall = recall_score(y_test, rf_predictions, average='weighted')
rf_f1 = f1_score(y_test, rf_predictions, average='weighted')

# Train and evaluate XGBoost
xgb_classifier = XGBClassifier(random_state=42)
xgb_classifier.fit(X_train, y_train)
xgb_predictions = xgb_classifier.predict(X_test)
xgb_accuracy = accuracy_score(y_test, xgb_predictions)
xgb_precision = precision_score(y_test, xgb_predictions, average='weighted')
xgb_recall = recall_score(y_test, xgb_predictions, average='weighted')
xgb_f1 = f1_score(y_test, xgb_predictions, average='weighted')

# Train and evaluate Support Vector Machine
svm_classifier = SVC(random_state=42)
svm_classifier.fit(X_train, y_train)
svm_predictions = svm_classifier.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_predictions)
svm_precision = precision_score(y_test, svm_predictions, average='weighted')
svm_recall = recall_score(y_test, svm_predictions, average='weighted')
svm_f1 = f1_score(y_test, svm_predictions, average='weighted')

# Print evaluation metrics for all models
print("Random Forest Classifier:")
print(classification_report(y_test, rf_predictions))
print("Accuracy:", rf_accuracy)
print("Precision:", rf_precision)
print("Recall:", rf_recall)
print("F1 Score:", rf_f1)
print("\n")

print("XGBoost Classifier:")
print(classification_report(y_test, xgb_predictions))
print("Accuracy:", xgb_accuracy)
print("Precision:", xgb_precision)
print("Recall:", xgb_recall)
print("F1 Score:", xgb_f1)
print("\n")

print("Support Vector Machine:")
print(classification_report(y_test, svm_predictions))
print("Accuracy:", svm_accuracy)
print("Precision:", svm_precision)
print("Recall:", svm_recall)
print("F1 Score:", svm_f1)
```

```python

# Define the models and their evaluation metrics
models = ['Random Forest', 'XGBoost', 'Support Vector Machine']
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
values = [
    [rf_accuracy, rf_precision, rf_recall, rf_f1],
    [xgb_accuracy, xgb_precision, xgb_recall, xgb_f1],
    [svm_accuracy, svm_precision, svm_recall, svm_f1]
]

# Plot the evaluation metrics for all models
plt.figure(figsize=(12, 8))

for i, metric in enumerate(metrics):
    plt.subplot(2, 2, i + 1)
    plt.bar(models, [values[j][i] for j in range(len(models))])
    plt.title(metric)
    plt.xlabel('Models')
    plt.ylabel(metric)
    plt.ylim(0, 1)  # Set y-axis limit to 0-1 for better visualization

plt.tight_layout()
plt.show()
```
